Bivariate plotting

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
reviews = pd.read_csv("data/winemag-data_first150k.csv", index_col=0)
reviews.head()
```

Scatterplot: 
maps each variable of interest to a point in two-dimensional space. 

```python
reviews[reviews['price'] < 100].sample(100).plot(kind='scatter', x='price', y='points')
```

price and points are weakly correlated: that is, that more expensive wines do generally earn more points when reviewed.

```python
reviews[reviews['price'] < 100].plot(kind='scatter', x='price', y='points')
```

Hex Plot

aggregates points in space into hexagons, and then colors those hexagons based on the values within them


```python
reviews[reviews['price'] <100].plot(kind='hexbin', x = 'price', y = 'points', gridsize=15)
```

Box plot: 
To visualize numerical data against a categorical data

```python
# box plots for the two columns
# reviews.loc[:,['price','points']].plot(kind='box')  #using .loc
# reviews[['price','points']].plot(kind='box')
reviews[(reviews.price < 100) & (reviews.points > 20)].plot(kind='box')
```

SUBPLOTTING



```python
fig, axarr = plt.subplots(2, 1, figsize=(12, 8))

reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0])

reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1])
```
