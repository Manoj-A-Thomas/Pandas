```python
import numpy as np
import pandas as pd
```

```python
url='https://raw.githubusercontent.com/Manoj-A-Thomas/data/data/diabetes-data-preprocessed.csv'
names=['preg','plas','pres','skin','test', 'mass', 
       'pedi','age','class']
dpp = pd.read_csv(url,names=names)
```

```python
dpp.shape
```

```python
dpp.info()
```

```python
dpp.sample(10)
```

```python
array = dpp.values

X = array[:, 0:8]
y = array[:,8]
```

```python
# X.shape
y.shape
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
# Feature selection - Filter method - chi square
test = SelectKBest(score_func=chi2, k=4)   # 4 most import target va
fit = test.fit(X,y)
```

```python
np.set_printoptions(precision=3)
print(fit.scores_)      #test, plas, mass, age
```

```python
# Scale features

features = fit.transform(X)
```

```python
print(features[0:5, :])     # sample from predictor dataset
```

```python
# Feature selection - Filter method - correlation coeff
features.dtype
features_df = pd.DataFrame(features)
features_df.corr()
```

```python
# Feature selection - Wrapper method - RFE

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
# Create the logistic regression model

model = LogisticRegression()
rfe = RFE(model, 3)     # chose top 3 featurs

fit = rfe.fit(X,y)
```

```python
print("selected features; %s" %(fit.support_))
print("Feature ranking: %s" %(fit.ranking_))

# feature to include: preg, pedi, age
```
